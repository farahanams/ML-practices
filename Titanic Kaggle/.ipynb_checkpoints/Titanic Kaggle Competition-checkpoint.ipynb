### 1. Problem Definition
We want to get the prediction of those who survived based on survival data in train data. The survived prediction is a binary value. It is safe to say that the problem is a classification rather than regression. There are many features with some missing-values. 

### 2. Define the success
If the Survived data of test data matched with predicted data, then, we are 100% succesful. However, our objective is to have 90% accuracy of prediction.

#### Structured data: 
There are many types of structured data. It depends on our interest, such as this, the binary data type of survived data indicates the goal. While name can be classified as text data. Some of the most important variables are: 

-- Categorical and numerical variables

Types of variables that can be classified (not exclusive for one type):
1. categorical variable: values of small set of possibilities.
2. numerical variable : discrete data and not continuous, grade of exam, # of children, siblings;
3. text variable: Name, Cabin, Embarked
4. binary variable: 0 and 1. 
5. relational variable: data that is related to each other, such as Pclass, cabin and ticket number.

** Number of siblings can be classified as numerical or categorical variable. It can be a small set of probabilities and discrete data which is not continuous.
